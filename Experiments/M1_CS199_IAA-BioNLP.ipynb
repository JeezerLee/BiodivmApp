```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score

# File path to INT_ALL_relations_summary
file_path = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/INT_ALL_relations_summary.xlsx"

old_xls = pd.ExcelFile(file_path)
df = pd.read_excel(old_xls, sheet_name="Sheet1")

# Compute overall percentage agreement
total_samples = len(df)
matching_labels = (df["relation_a"] == df["relation_b"]).sum()
overall_agreement = (matching_labels / total_samples) * 100


labels = ["has_taxon", "has_habitat", "no_relation", "x"]
label_agreement = {
    label: (df[(df["relation_a"] == label) & (df["relation_b"] == label)].shape[0] / df[df["relation_a"] == label].shape[0]) * 100
    if df[df["relation_a"] == label].shape[0] > 0 else 0
    for label in labels
}

# Compute Cohen's Kappa
kappa = cohen_kappa_score(df["relation_a"], df["relation_b"])

# Print results
print(f"Overall Agreement: {overall_agreement:.2f}%")
print("Agreement per Label:")
for label, agreement in label_agreement.items():
    print(f"  {label}: {agreement:.2f}%")
print(f"Cohen's Kappa: {kappa:.2f}")
```
