```python
import pandas as pd

# Define paths for train, dev, and test datasets in Google Drive
train_path_v1 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/f_train.xlsx"
dev_path_v1 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/f_dev.xlsx"
test_path_v1 ="/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/f_test.xlsx"
train_path_v2 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/g_train.xlsx"
dev_path_v2 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/g_dev.xlsx"
test_path_v2 ="/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/g_test.xlsx"

# Load the Excel files
train_df_v1 = pd.read_excel(train_path_v1)
dev_df_v1 = pd.read_excel(dev_path_v1)
test_df_v1 = pd.read_excel(test_path_v1)

train_df_v2 = pd.read_excel(train_path_v2)
dev_df_v2 = pd.read_excel(dev_path_v2)
test_df_v2 = pd.read_excel(test_path_v2)

print("Loaded")
```

```python
def preprocess_data(df):
    processed_inputs = []

    for chunk, entity1, entity2, entity_type1, entity_type2 in zip(
        df["Chunk"], df["Entity 1"], df["Entity 2"], df["Entity 1 Type"], df["Entity 2 Type"]
    ):
        # Replace entities with their types
        # print(entity1, entity_type1, entity2, entity_type2)
        input_text = chunk.replace(entity1, f"[{entity_type1}]").replace(entity2, f"[{entity_type2}]")
        processed_inputs.append(input_text)

    # Create a new dataframe with only `chunk` (processed text) and `relation`
    df_processed = pd.DataFrame({
        "chunk": processed_inputs,  # Preprocessed text
        "relation": df["relation"]  # Relation labels
    })

    return df_processed


train_processed_v1 = preprocess_data(train_df_v1)
dev_processed_v1 = preprocess_data(dev_df_v1)
test_processed_v1 = preprocess_data(test_df_v1)

train_processed_v2 = preprocess_data(train_df_v2)
dev_processed_v2 = preprocess_data(dev_df_v2)
test_processed_v2 = preprocess_data(test_df_v2)


print("Done")

print("\n### Sample Preprocessed Sentences from the training Set ###")
display(train_processed_v2.head(5))
```

```python
pre_train_path_v1 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/MASKED/MPREMASK_TRAIN_V1.xlsx"
pre_dev_path_v1 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/MASKED/MPREMASK_DEV_V1.xlsx"
pre_test_path_v1 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/MASKED/MPREMASK_TEST_V1.xlsx"

pre_train_path_v2 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/MASKED/MPREMASK_TRAIN_V2.xlsx"
pre_dev_path_v2 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/MASKED/MPREMASK_DEV_V2.xlsx"
pre_test_path_v2 = "/content/drive/MyDrive/2425_2NDSEM/CS199/manuscript/MDATA/MASKED/MPREMASK_TEST_V2.xlsx"

# Save preprocessed datasets with only `chunk` and `relation`
train_processed_v1.to_excel(pre_train_path_v1, index=False)
dev_processed_v1.to_excel(pre_dev_path_v1, index=False)
test_processed_v1.to_excel(pre_test_path_v1, index=False)

train_processed_v2.to_excel(pre_train_path_v2, index=False)
dev_processed_v2.to_excel(pre_dev_path_v2, index=False)
test_processed_v2.to_excel(pre_test_path_v2, index=False)

print("Preprocessed datasets saved")
```
